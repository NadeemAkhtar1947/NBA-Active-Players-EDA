```python
country_distribution = df['country'].value_counts()
plt.figure(figsize=(12,  12))
country_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=360, colors=plt.cm.Paired.colors)
# Create a pie chart
plt.title('Distribution of country',color='indigo',size=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python
df['Human Development Groups'].values
```

```python
df['Human Development Groups'].value_counts()
```

```python
country_distribution_by_HDG = df['Human Development Groups'].value_counts()
fig, ax = plt.subplots(figsize=(16, 5))
country_distribution_by_HDG.plot(kind='bar', color='indigo')
plt.ylabel('Count', color='green',size=20)
plt.xlabel('Country', color='blue',size=20)
plt.title('Distribution of country by Human Development Groups', color='red',size=22)

plt.xticks(rotation=90) #Rotate the country names for better readability 
                        # Note:- if you won't provide rotation then it automatically takes 90 degree

plt.show()
```

```python
country_distribution_by_HDG = df['Human Development Groups'].value_counts()
plt.figure(figsize=(12,  12))
country_distribution_by_HDG.plot(kind='pie', autopct='%1.1f%%', startangle=360, colors=plt.cm.Paired.colors)

# Create a pie chart
plt.title('Distribution of country by Human Development Groups',color='indigo',size=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```
